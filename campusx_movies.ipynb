```python
import pandas as pd

# 1. Read the problematic CSV using a flexible encoding
# 'latin-1' (or 'ISO-8859-1') is very lenient and avoids the 'invalid byte sequence' error.
# We must specify the correct delimiter (;) from your file.
try:
    df = pd.read_csv('movies.csv', delimiter=';', encoding='latin-1')
except UnicodeDecodeError as e:
    print(f"Could not read with latin-1: {e}")
    # Fallback plan if latin-1 fails (though highly unlikely for this error type)
    exit()

# 2. Write the DataFrame back to a new CSV with the required UTF-8 encoding
# This step forces every character to be correctly represented in UTF-8 format.
output_file = 'movies_fixed_utf8.csv'
df.to_csv(output_file, index=False, sep=';', encoding='utf-8')

print(f"Successfully converted and saved the file as: {output_file}")
```
